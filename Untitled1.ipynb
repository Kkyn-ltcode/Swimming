```python
def score(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    return {'accuracy': accuracy, 'precision': precision,
            'recall': recall, 'f1_score': f1}
```

```python
r = export_text(estimator, feature_names = train_data.columns.to_list(), max_depth=100)
rule = []
features = []
values = []
num = 1
encode_cols = {col: letter for col, letter in zip(train_data.columns, string.ascii_uppercase)}
pink_eye_file = open(f"data/{out_name}.txt", "w")
for i in r.split('\n')[:-1]:
    i = i.split('|--- ')[-1]
    if 'class' in i:
        label = int(float(i.split()[-1]))
        result = f'Rule {num}: ' + '| '.join(rule) + f' --> {label}'
        pink_eye_file.write(result + '\n')
        num += 1
        print(result)
    else:
        feature = i.split()[0]
        value = i.split()[-1]
        i = i.replace(feature, encode_cols[feature])
        if feature not in ['age', 'swimming_time']:
            i = i.replace('<= 0.50', '= 0')
            i = i.replace('>  0.50', '= 1')
        else:
            i = i.replace('<  ', '< ')
            i = i.replace('>  ', '> ')
        if feature in features:
            f = features[::-1]
            index = len(f) - 1 - f.index(feature)
            if values[index] == value:
                del rule[index:]
                del features[index:]
                del values[index:]
        rule.append(i)
        features.append(feature)
        values.append(value)
pink_eye_file.close()
```
